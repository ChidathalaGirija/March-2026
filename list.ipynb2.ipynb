# combining course_list,student_count lists using Zip

```python
course_list = ["data engineering","data analytics","ml engineering","powerbi developer","azure devops","java developet","sql fundamentals"]
student_count = [120,220,320,420,520,620,720]
for course_list,student_count in zip(course_list, student_count):
      print(course_list ,"-", student_count)
```
