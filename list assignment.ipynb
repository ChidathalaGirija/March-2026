Question 10: Create a list of prime numbers between 1 and 50

```python
primes =[]
for num in range(2,51):
    count = 0
    for i in range (1,num+1):
        if num % i==0:
            count +=1
    if count == 2:
        primes.append(num)
print(primes)
```

Question 18: Find the sum of all elements in 3D list [[[1, 2], [3, 4]], [[5, 6], [7, 8]]]

```python
list_3d = [[[1, 2], [3, 4]], [[5, 6], [7, 8]]]
total = 0
for sub1 in list_3d:
    for sub2 in sub1:
        for num in sub2:
            total += num
print(total)
```

Question 19: Extract all even numbers from nested list [[1, 2, 3], [4, 5, 6], [7, 8, 9]]

```python
nested = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
evens = []
for sublist in nested:
    for num in sublist:
        if num % 2 ==0:
            evens.append(num)
print(evens)
```
